```python
# g
fighters = mlbb_data[mlbb_data['role'] == 'fighter']
tanks = mlbb_data[mlbb_data['role'] == 'tank']

avg_physical_attack_fighters = fighters['physical_atk'].mean()
avg_physical_attack_tanks = tanks['physical_atk'].mean()

print("Average physical attack for fighters:", avg_physical_attack_fighters)
print("Average physical attack for tanks:", avg_physical_attack_tanks)

if(avg_physical_attack_fighters > avg_physical_attack_tanks):
  print("Fighters have higher average physical attack than Tanks.")
elif (avg_physical_attack_fighters < avg_physical_attack_tanks):
  print("Tanks have higher average physical attack than Fighters.")
else:
  print("Fighters and Tanks have the same average physical attack.")
```

7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
# 7

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Encode the 'role' column
label_encoder = LabelEncoder()
mlbb_data['role_encoded'] = label_encoder.fit_transform(mlbb_data['role'])

# Separate features and target variable
x = mlbb_data[['defense_overall','offense_overall','skill_effect_overall','difficulty_overall','movement_spd','magic_defense','mana','hp_regen','physical_atk','physical_defense','hp','attack_speed','mana_regen']]

y = mlbb_data['role_encoded']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)

# Create a Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the model
clf.fit(x_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of the model:", accuracy)

# Predict the role for the given character
predicted_role_encoded = clf.predict([[6.0,7.0,4.0,3.0,250,9,380,37,119,24,2689,0.75,16.0]])

# Get the predicted role (dicoding)
predicted_role = label_encoder.inverse_transform(predicted_role_encoded)

print(f"Predicted role for the character: {predicted_role[0]}")
```
